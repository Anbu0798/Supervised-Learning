```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/aps1070-2019/datasets/master/LR_data.csv" , skipinitialspace=True)
```

```python
df
```

```python
import numpy as np
column = list(df.columns)
column_y = column.pop(-1)
column_X = column
column_X
```

```python
X = df[column_X].to_numpy()
y = df[column_y].to_numpy()
X
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = 0, 0, 0, 0
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, random_state=1)
X_train.shape
```

```python
# Standaradizing the data
from sklearn.preprocessing import StandardScaler
scal = StandardScaler()
X_train_std = scal.fit_transform(X_train)
X_val_std = scal.transform(X_val)
```

```python
def RMSE(x, y):
  ans = sqrt(mean_squared_error(x, y))
  return ans
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
mdl = LinearRegression().fit(X_train_std, y_train)

y_pred_train = mdl.predict(X_train_std)
y_pred_val = mdl.predict(X_val_std)
RMSE_train = RMSE(y_pred_train, y_train)

# RMSE for training and validation sets
RMSE_val = RMSE(y_pred_val, y_val)
print("RMSE for training:", RMSE_train)
print("RMSE for validation:", RMSE_val)
```

```python
# Linear Regression with Additional Features
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=8)
X_poly_train = poly.fit_transform(X_train)
X_poly_val = poly.transform(X_val)

# Standardizing
scaler = StandardScaler()
X_poly_train_std = scaler.fit_transform(X_poly_train)
X_poly_val_std = scaler.transform(X_poly_val)

model = LinearRegression().fit(X_poly_train_std, y_train)
y_pred_poly_train = model.predict(X_poly_train_std)
y_pred_poly_val = model.predict(X_poly_val_std)

RMSE_poly_train = RMSE(y_pred_poly_train, y_train)
RMSE_poly_val = RMSE(y_pred_poly_val, y_val)
print("RMSE for training with additional features:", RMSE_poly_train)
print("RMSE for validation with additional features:", RMSE_poly_val)
```

```python
# Linear Regression with Additional Features and Regularization
from sklearn.linear_model import Ridge
RMSE_reg_train = []
RMSE_reg_val = []
alpha = [10**i for i in range(-2,11)]
alpha
for i in alpha:
  model_reg = Ridge(alpha=i).fit(X_poly_train_std, y_train)
  y_pred_reg_train = model_reg.predict(X_poly_train_std)
  y_pred_reg_val = model_reg.predict(X_poly_val_std)
  RMSE_reg_train.append(RMSE(y_pred_reg_train, y_train))
  RMSE_reg_val.append(RMSE(y_pred_reg_val, y_val))
df_table = pd.DataFrame(list(zip(alpha, RMSE_reg_train, RMSE_reg_val)), columns=['Alpha', 'Training RMSE', 'Validation RMSE'])
df_table
```
